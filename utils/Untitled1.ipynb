```python
from pathlib import Path
import sys
sys.setrecursionlimit = 10000
```

```python
help_path = Path('/home/neum/Documenti/SuperCollider/HelpSource')
```

```python
def iter_schelp(p, coll):
    for file in p.iterdir():
        if file.is_dir():
            iter_schelp(file, coll)
        else:
            if file.suffix == '.schelp':
                coll.append(file)
```

```python
schelps = []
iter_schelp(help_path, schelps)
```
